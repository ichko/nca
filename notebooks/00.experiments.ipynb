```python
from nca.models.basic_nca import BasicNCA
import numpy as np
```

```python
# init = torch.rand(4, 1, 40, 40) > 0.7
init = torch.rand(4, 1, 40, 40)

pos_x, pos_y = 30, 30
h, w = init.shape[-2:]
start[:, :, pos_y:pos_y+h, pos_x:pos_x+w] = init
```

```python
nca = BasicNCA()
```

```python
x_space = torch.linspace(-1, 1, 1000).unsqueeze(-1)

y_map = nca.rule(x_space)

np_x = x_space.ae.np
np_y = y_map.ae.np
```

```python
plt.plot(np_x[:,0], np_y[:,0])
```

```python
seq = nca(start, steps=300)
out = torch.stack(seq, axis=1)[:, :, 0]
out.shape
```
